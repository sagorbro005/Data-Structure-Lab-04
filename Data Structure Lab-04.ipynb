```python
class Patient:
  def __init__(self,id,name,age,bloodgroup,next=None,prev=None):
    self.id=id
    self.name=name
    self.age=age
    self.bloodgroup=bloodgroup
    self.next=next
    self.prev=prev
class WRM:
  def __init__(self):
    self.dh=Patient(None,None,None,None)
    self.dh.next=self.dh
    self.dh.prev=self.dh
  def RegisterPatient(self,id,name,age,bloodgroup):
    newPatient=Patient(id,name,age,bloodgroup)
    tail=self.dh
    tail.next=newPatient
    newPatient.prev=tail
    newPatient.next=self.dh
    self.dh.prev=newPatient
  def ServePatient(self):
    if self.dh==self.dh.next:
      print("No Patient to serve")
    else:
      serve_patient=self.dh.next
      serve_patient.next.prev=self.dh
      self.dh.next=serve_patient.next
      print(f"Patient to serve: {serve_patient.name}")
  def CancelAll(self):
    if self.dh.next==self.dh:
      print("There is no patient to cancel")
    else:
      self.dh.next=self.dh
      self.dh.prev=self.dh
  def CanDoctorGoHome(self):
    if self.dh.next==self.dh:
      print("Yes")
    else:
      print("No")
  def ShowAllPatient(self):
    current_patient=self.dh.next
    while current_patient!=self.dh:
      print(current_patient.id)
      current_patient=current_patient.next
wrm=WRM()
while True:
    print("Welcome to BRAC Hospital.Choose any option below:")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All patients")
    print("4. Can Doctor go Home?")
    print("5. Cancel all Appointment")
    print("6. Exit")
    option=input("Choose an option: ")
    if option=="1":
      id=int(input("Enter Patient ID: "))
      name=input("Enter Patient Name: ")
      age=int(input("Enter Patient Age: "))
      bloodgroup=input("Enter Patient Blood Group: ")
      wrm.RegisterPatient(id,name,age,bloodgroup)
      print("Success")
    elif option=="2":
      wrm.ServePatient()
    elif option=="3":
      wrm.ShowAllPatient()
    elif option=="4":
      wrm.CanDoctorGoHome()
    elif option=="5":
      wrm.CancelAll()
      print("Success")
    elif option=="6":
      print("Thank You")
      break
```
